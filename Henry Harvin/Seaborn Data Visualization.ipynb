## Seaborn

Seaborn is a data visualization library built on top of matplotlib. It has a high-level interface as compared to matplotlib.

### Visualizing Statistical Relationships

### Scatter Plot

A scatter plot is used to define the relationship between two variables. It can be used to find if by increasing value of one variable leads to an increase in another variable and vice versa.

```python
sns.relplot(x = 'horsepower',y = 'price', data = df)
```

```python
# Tag associated with the data
sns.relplot(x='horsepower',y='price',hue='body_style',data=df)
```

In the above plot, apart from getting the relationship between horsepower(x) and price(y), we can also see another visualization using Hue. This is called Hue Plot 

### Plotting Categorical Data

### Jitter Plot

```python
sns.catplot(x='body_style',y='price',data=df)
```

Since the plot is scattered, we can handle it using jitter as false. Jitter is the deviation from the true value

```python
sns.catplot(x='body_style',y='price',jitter=False,data=df)
```

We can also add hue to the plots similar to the one with scatter plots.

```python
sns.catplot(x='body_style',y='price',hue='fuel_type',data=df)
```

We can see the points are overlapping here. To reduce the overlapping, we can use swarm algorithm (using kind='swarm').

```python
sns.catplot(x='body_style',y='price',kind='swarm',data=df)
```

```python
sns.catplot(x='body_style',y='price',kind='swarm',hue='fuel_type',data=df)
```

### Box Plot

```python
sns.catplot(x='body_style',y='price',kind='box',data=df)
```

### Violin Plot

Violin Plots combine box plots and kernel distribution to provide a better insight into the distribution of values.

```python
sns.catplot(x='body_style',y='price',kind='violin',data=df)
```

### Point Plot

### HeatMap

### Boxen Plot

### Ridge/Joy Plots 

```python
g = sns.FacetGrid(df,row='body_style',hue='fuel_type',aspect=5,height=3)
g.map(sns.kdeplot, "price", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "price", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "price")
g.fig.subplots_adjust(hspace=-.25)
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
